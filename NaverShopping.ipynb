```python
import urllib.request

client_id = '3491KcqXLTJvq3HFRt63'
client_secret = 'ykgUcEDsya'

site_url = 	'https://openapi.naver.com/v1/search/shop.json'
```

```python
query = input("검색어:")

q_param = "query="+urllib.parse.quote(query)

display_num = int(input("1~100 : "))
display_param ="display=%d"%display_num
cid = 'X-Naver-Client-Id'
csec = 'X-Naver-Client-Secret'

query_str = site_url+'?'+q_param+'&'+display_param
print(query_str)

request = urllib.request.Request(query_str) 
request.add_header(cid,client_id) 
request.add_header(csec,client_secret)

response = urllib.request.urlopen(request) #웹 서버에 요청
if response.getcode() != 200:
  print("실패")
else:
  print('성공')
```

```python
data = response.read().decode("utf-8")
print(data)
```

```python
import json
```

```python
jdata = json.loads(data)
items = jdata['items']
print(items)
```

```python
for item in items:
  print("_______________")
  print("제목:",item['title'])
  print("가격",item['lprice'])
  print("이미지",item['image'])
```
